```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
maindata = pd.read_csv('Data_for_Transformation.csv')
plt.scatter(maindata["Age"], maindata["Salary"])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
plt.hist(maindata["Salary"], bins = 10, color = "green")
plt.xlabel("Salaries")
plt.show()
```

```python
size = plt.figure(figsize=(7, 5))
plt.bar(maindata["Country"], maindata["Salary"], color="blue")
plt.xlabel("Salaries")
plt.ylabel("Countries")
plt.title("Bar chart of Country vs Salary")
plt.show()

```
